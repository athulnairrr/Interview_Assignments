Data Loading / Data Understanding

Load the Dataset first

I have divided the Dataset as 3 parts:

1. Skills['Python (out of 3)', 'Machine Learning (out of 3)','Natural Language Processing (NLP) (out of 3)','Deep Learning (out of 3)', 'Other skills',]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Data
file = '/content/Applications_for_Machine_Learning_internship_edited.xlsx'
data = pd.read_excel(file)
```

```python
data.head(10)
```

```python
total_data = data.count()
total_data
```

```python
# Education section due to having so much null value needed to be removed

missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Forming a new Updated Dataset

data_updated = data.iloc[:, 0:7]
data_updated
```

```python
# Updating the value for availability as 1 if in the string yes exists otherwise 0

available = 'yes'
for available in data_updated['Are you available for 3 months, starting immediately, for a full-time work from home internship? ']:
    if isinstance(available, str):
      data_updated['Are you available for 3 months, starting immediately, for a full-time work from home internship? '] = 1
    else:
      data_updated['Are you available for 3 months, starting immediately, for a full-time work from home internship? '] = 0

data_updated.head()
```

```python
######
```

The section containing Other skills was used to determine the various other skill sets that an intern can bring to the table

Based on Role Requiremnets for an intern i personally gave certain skills a ranking which was then used in the combined score to ensure the best intern is selected

```python


other_skills = []
for skills in data_updated['Other skills']:
    if isinstance(skills, str):
        skills_list = skills.split(', ')
        other_skills.extend(skills_list)
```

```python
distinct_skills = pd.Series(other_skills).value_counts().head(20)
print(distinct_skills)
```

```python
skill_weights = {
    'Data Science': 4,
    'Data Analytics': 3,
    'SQL': 2,
    'MS-Excel': 1,
   }

data_updated['Skill Score'] = data_updated['Other skills'].apply(lambda skills: sum(skill_weights.get(skill, 0) for skill in skills.split(', ')) if pd.notnull(skills) and not isinstance(skills, int) else 0)
data_updated.drop(['Other skills'], axis=1, inplace = True)
data_updated.head()
```

```python
data_updated['combined_score'] = data_updated[['Python (out of 3)', 'Machine Learning (out of 3)',
       'Natural Language Processing (NLP) (out of 3)',
       'Deep Learning (out of 3)',
       'Are you available for 3 months, starting immediately, for a full-time work from home internship? ', 'Skill Score']].sum(axis=1)
Best_Interns = data_updated['combined_score'].max()
Best_Interns
```

```python
######
```

Based on the final score the ones with the maximum score are separated into another dataset that can be then used to easily identify the Best Interns

In this case we have around 6 Best Interns where each of them would provide the same value to the company

For further finding the best these 6 can be interviewed to get an insight into their soft skills to better understnad who would be the best fit for the Company

```python
Best_Interns_rows = data_updated.loc[data_updated['combined_score'] == Best_Interns]
Best_Interns_rows.head(10)
```
